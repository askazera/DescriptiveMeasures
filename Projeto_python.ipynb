-- Média aritmética;
-- Mediana;

-- Variância amostral;
-- Desvio Padrão amostral;
-- Amplitude total;
-- Q1 (primeiro quartil);
-- Q3 (terceiro quartil);
-- AIQ (amplitude interquartílica);
-- Limite inferior (LI do boxplot);
-- Limite superior (LS do boxplot);
-- Identificação e classificação de possíveis outliers na amostra;
-- Coeficiente de Correlação de Pearson ( r );
-- Coeficiente de determinação ( r ao quadrado);
-- Coeficiente a ou alpha da equação de regressão linear simples para as duas -- variáveis escolhidas pelo usuário;
-- Coeficiente b ou beta da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário;



```python
while True:
    valor1 = input('Informe uma sequência de números desejada:')
    valor2 = input('Informe uma segunda sequência de números desejada:')
    valor1 = valor1.split() # separando os valores dentro de uma lista
    valor2 = valor2.split()

    try:
        lista1 = [float(i) for i in valor1] # transformando cada item de string para float e colocando em lista
        lista2 = [float(i) for i in valor2]
        if len(lista1) == len(lista2):
            break
        else:
            print("As sequências devem ter o mesmo tamanho. Por favor, tente novamente.")
    except ValueError:
        print("Os valores digitados devem ser numéricos. Por favor, tente novamente.")


# Restante do código...

# 1. MÉDIA ARITMÉTICA

media1 = sum(lista1) / len(lista1)
media2 = sum(lista2) / len(lista2)
print(f'\nA média aritmética da primeira sequência de números é: {media1:.2f}')
print(f'A média aritmética da segunda sequência de números é: {media2:.2f}')


# 2. MEDIANA

# colocando os valores em ordem

ordem1 = sorted(lista1)
ordem2 = sorted(lista2)

# achando a mediana da primeira lista
if len(ordem1) % 2 == 0:
    med1= len(ordem1)//2
    med2 = len(ordem1)//2-1
    mediana = (ordem1[med1]+ordem1[med2])/2
    print(f'\nA mediana da primeira sequência de números é: {mediana}')
else:
    med= len(ordem1)//2
    mediana = ordem1[med]
    print(f'\nA mediana da primeira sequência de números é: {mediana}')

# achando a mediana da segunda lista
if len(ordem2) % 2 == 0:
    med1= len(ordem2)//2
    med2 = len(ordem2)//2-1
    mediana = (ordem2[med1]+ordem2[med2])/2
    print(f'A mediana da segunda sequência de números é: {mediana}')
else:
    med= len(ordem2)//2
    mediana = ordem2[med]
    print(f'A mediana da segunda sequência de números é: {mediana}')


# 3. MODA

moda = []
frequencias = []
for i in lista1:  # Calcula a frequência de cada número
    frequencia = 0
    for x in lista1:
        if i == x:
            frequencia += 1
    frequencias.append(frequencia)

maior_frequencia = 0

for i in range(len(frequencias)):  # Encontra a maior frequência
    if frequencias[i] > maior_frequencia:
        maior_frequencia = frequencias[i]
        moda = [lista1[i]]
    elif frequencias[i] == maior_frequencia and lista1[i] not in moda:
        moda.append(lista1[i])

# Exibe os números com a maior frequência na tela
if len(moda) == 1:
    print(f'\nA moda da lista 1 é: {moda[0]}')
elif len(moda) != len(frequencias):
    print(f'\nAs modas da lista 1 são: {moda}')
else:
    print('\nA lista 1 não possui moda.')

# Moda da lista 2

moda = []
frequencias = []

for i in lista2:  # Calcula a frequência de cada número
    frequencia = 0
    for x in lista2:
        if i == x:
            frequencia += 1
    frequencias.append(frequencia)

maior_frequencia = 0

for i in range(len(frequencias)):  # Encontra a maior frequência
    if frequencias[i] > maior_frequencia:
        maior_frequencia = frequencias[i]
        moda = [lista2[i]]
    elif frequencias[i] == maior_frequencia and lista2[i] not in moda:
        moda.append(lista2[i])

# Exibe os números com a maior frequência na tela
if len(moda) == 1:
    print(f'A moda da lista 2 é: {moda[0]}')
elif len(moda) != len(frequencias):
    print(f'As modas da lista 2 são: {moda}')
else:
    print('A lista 2 não possui moda.')


# 4. VARIÂNCIA

# calculo da variância amostral para dados quantitativos, por isso -1 ao final do cálculo

variancia_1 = sum((i - media1) ** 2 for i in lista1) / (len(lista1) -1)
variancia_2 = sum((i - media2) ** 2 for i in lista2) / (len(lista2) -1)


print(f'\nA variância da lista 1 é: {variancia_1:.2f}')
print(f'A variância da lista 2 é: {variancia_2:.2f}')


# 5. DESVIO PADRÃO


dp1 = variancia_1**0.5
dp2 = variancia_2**0.5

print(f'\nO desvio padrão da lista 1 é: {dp1:.2f}')
print(f'O desvio padrão da lista 2 é: {dp2:.2f}')


# 6. AMPLITUDE TOTAL


at1 = (ordem1[-1] - ordem1[0]) # ultimo valor da lista - primeiro valor da lista
at2 = (ordem2[-1] - ordem2[0])


print(f'\nA amplitude total da lista 1 é: {at1}')
print(f'A amplitude total da lista 2 é: {at2}')


# 7. Q1 (PRIMEIRO QUARTIL)

ordem1 = sorted(lista1)
ordem2 = sorted(lista2)

# achando o primeiro quartil de uma sequência par

if len(ordem1) % 2 == 0: # se a quantidade de números for par
    qua1_1= (len(ordem1)+1)//4
    print(f'\nO primeiro quartil da lista 1 é: {ordem1[qua1_1]}')


# encontrando o primeiro quartil de uma sequência ímpar

else:
    med= len(ordem1)//2 # se a quantidade de números for ímpar , faz esse calculo para achar a mediana na lista
    mediana = ordem1[med] # achamos a mediana da lista ordenada
    q1_1 = ordem1[:med]   # achando a lista de primeiro quartil Q1
    try:
        if len(q1_1) % 2 == 0: # se a quantidade de números do Q1 for par
            me1= len(q1_1)//2
            me2= len(q1_1)//2-1
            qua1_1 = (q1_1[me1]+q1_1[me2])/2
            print(f'\nO primeiro quartil da lista 1 é: {qua1_1}')
    except:
        print('\nNão há primeiro quartil na lista 1')


# Q1 lista 2

# se o tamanho da lista for par

if len(ordem2) % 2 == 0:
    qua1_2 = (len(ordem2)+1)//4
    print(f'O primeiro quartil da lista 2 é: {ordem2[qua1_2]}')


# encontrando o primeiro quartil de uma sequência ímpar
else:
    med= len(ordem2)//2 # se a quantidade de números for ímpar , faz esse calculo para achar a mediana na lista
    mediana = ordem2[med] # achamos a mediana da lista ordenada, indice
    q2_2 = ordem2[:med]   # achando o intervalo de primeiro quartil Q1
    try:
        if len(q2_2) % 2 == 0: # se a quantidade de números do Q1 for par
            me1= len(q2_2)//2
            me2= len(q2_2)//2-1
            qua1_2 = (q2_2[me1]+q2_2[me2])/2
            print(f'O primeiro quartil da lista 2 é: {qua1_2}')
    except:
        print('Não há primeiro quartil na lista 2')


# 8. Q3 (TERCEIRO QUARTIL)


ordem1 = sorted(lista1)
ordem2 = sorted(lista2)

# Q3 da lista 1


if len(ordem1) % 2 == 0: # se a quantidade de números for par
    qua3_1= (3*(len(ordem1)+1))//4
    print(f'\nO terceiro quartil da lista 1 é: {ordem1[qua3_1]}')

else:
    med= len(ordem1)//2 # se a quantidade de números for ímpar , faz esse calculo para achar a mediana na lista
    mediana = ordem1[med] # achamos a mediana da lista ordenada
    q3_1 = ordem1[med+1:]   # achando a lista de terceiro quartil Q3, pegando o intervalo após a mediana
    try:
        if len(q3_1) % 2 == 0: # se a quantidade de números do Q3 for par
            me1= len(q3_1)//2
            me2= len(q3_1)//2-1
            qua3_1 = (q3_1[me1]+q3_1[me2])/2
            print(f'\nO valor do terceiro quartil da lista 1 é: {qua3_1}')
    except:
        print('\nNão há primeiro quartil para a lista 1')


# Q3 da lista 2


if len(ordem2) % 2 == 0: # se a quantidade de números for par
    qua3_2 = (3*(len(ordem2)+1))//4
    print(f'O terceiro quartil da lista 1 é: {ordem2[qua3_2]}')

else:
    med= len(ordem2)//2 # se a quantidade de números for ímpar , faz esse calculo para achar a mediana na lista
    mediana = ordem2[med] # achamos a mediana da lista ordenada
    q3_2 = ordem2[med+1:]   # achando a lista de terceiro quartil Q3, pegando o intervalo após a mediana
    try:
        if len(q3_2) % 2 == 0: # se a quantidade de números do Q3 for par
            me1= len(q3_2)//2
            me2= len(q3_2)//2-1
            qua3_2 = (q3_2[me1]+q3_2[me2])/2
            print(f'O valor do terceiro quatil da lista 2 é: {qua3_2}')
    except:
        print('Não há terceiro quartil para a lista 2')


# 9. AIQ/IQR (amplitude interquartílica)

# AI1 = Q3 - Q1

if len(ordem1) % 2 == 0: # caso o tamanho da sequência seja par
    aiq_1 = ordem1[qua3_1] - ordem1[qua1_1]
    print(f'\nAmplitude interquatílica da lista 1: {aiq_1}')
else: # se o tamanho da sequência for ímpar
    aiq_1 = qua3_1 - qua1_1
    print(f'\nAmplitude interquatílica da lista 1: {aiq_1}')

if len(ordem2) % 2 == 0:
    aiq_2 = ordem2[qua3_2] - ordem2[qua1_2]
    print(f'Amplitude interquatílica da lista 2: {aiq_2}')
else:
    aiq_2 = qua3_2 - qua1_2
    print(f'Amplitude interquatílica da lista 2: {aiq_2}')



# 10. LIMITE INFERIOR (LI)

# LI = Q1 - 1.5 *(AIQ)

if len(ordem1) % 2 == 0: # caso o tamanho da sequência seja par
    li1 = ordem1[qua1_1] - 1.5*(aiq_1)
    print(f'\nO limite inferior da lista 1 é: {li1}')
else: # se o tamanho da sequência for ímpar
    li1 = qua1_1 - 1.5*(aiq_1)
    print(f'\nO limite inferior da lista 1 é: {li1}')

if len(ordem2) % 2 == 0:
    li2 = ordem2[qua1_2] - 1.5*(aiq_2)
    print(f'O limite inferior da lista 2: {li2}')
else:
    li2 = qua1_2 - 1.5*(aiq_2)
    print(f'O limite inferior da lista 2: {li2}')


# 11. LIMITE SUPERIOR (LS)

# LS = Q3 + 1.5*(AIQ)

if len(ordem1) % 2 == 0: # caso o tamanho da sequência seja par
    ls1 = ordem1[qua3_1] + 1.5*(aiq_1)
    print(f'\nO limite inferior da lista 1 é: {ls1}')
else: # se o tamanho da sequência for ímpar
    ls1 = qua3_1 + 1.5*(aiq_1)
    print(f'\nO limite superior da lista 1 é: {ls1}')

if len(ordem2) % 2 == 0:
    ls2 = ordem2[qua3_2] + 1.5*(aiq_2)
    print(f'O limite inferior da lista 2: {ls2}')
else:
    ls2 = qua3_2 + 1.5*(aiq_2)
    print(f'o limite superior da lista 2: {ls2}')


# 12. OUTLIERS

# outlier lista 1


out_sup1 = [] # lista para inserir os outliers do limite superior da primeira lista
out_inf1 = [] #  lista para inserir os outliers do limite inferior da primeira lista

for i in ordem1: # para cada valor na lista ordenada
    if i > ls1 and i not in out_sup1: # se o valor for maior que o limite superior e nao estiver na lista
        out_sup1.append(i)
    if i < li1 and i not in out_inf1: # se o valor for menor que o limite inferior e nao estiver na lista
        out_inf1.append(i)

if len(out_sup1) == 1:
    print(f'\nO outlier superior da lista 1 é: {out_sup1}')
elif len(out_sup1) > 1:
    print(f'\nOs outliers superiores da lista 1 é: {out_sup1}')
else:
    print(f'\nNão há outlier superior na lista 1')

if len(out_inf1) == 1:
    print(f'\nO outlier inferior da lista 1 é: {out_inf1}')
elif len(out_inf1) > 1:
    print(f'\nOs outliers inferiores da lista 1 é: {out_inf1}')
else:
    print(f'Não há outlier inferior na lista 1')


# outlier lista 2


out_sup2 = []  #lista para inserir os outliers do limite superior da segunda lista
out_inf2 = []  #lista para inserir os outliers do limite inferior da segunda lista

for i in ordem2:
    if i > ls2 and i not in out_sup2:
        out_sup2.append(i)
    if i < li2 and i not in out_inf2:
        out_inf2.append(i)

if len(out_sup2) == 1:
    print(f'\nO outlier superior da lista 2 é: {out_sup2}')
elif len(out_sup2) > 1:
    print(f'\nOs outliers superiores da lista 2 são: {out_sup2}')
else:
    print(f'Não há outlier superior na lista 2')

if len(out_inf2) == 1:
    print(f'\nO outlier inferior da lista 2 é: {out_inf2}')
elif len(out_inf2) > 1:
    print(f'\nOs outliers inferiores da lista 2 são: {out_inf2}')
else:
    print(f'Não há outlier inferior na lista 2')


# 13. COEFICIENTE DE CORRELAÇÃO DE PEARSON (r)


tamanho_l1 = len(lista1)
tamanho_l2 = len(lista2)

media1 = sum(lista1) / len(lista1)
media2 = sum(lista2) / len(lista2)

# desvio padrão
dp1 = variancia_1**0.5
dp2 = variancia_2**0.5

numerador = 0
denominador_x = 0
denominador_y = 0


for i in range(tamanho_l1):
    numerador += (lista1[i] - media1) * (lista2[i] - media2)
    denominador_x += (lista1[i] - media1) ** 2
    denominador_y += (lista2[i] - media2) ** 2

coef_pearson = numerador / ((denominador_x ** 0.5) * (denominador_y ** 0.5))
print(f'\nCoeficiente de correlação de Pearson: {coef_pearson:.2f}')


# 14. COEFICIENTE DE DETERMINAÇÃO ( r ao quadrado)

print(f'\nCoeficiente de determinação: {(coef_pearson)**2:.2f}')



# 15. COEFICIENTE B OU BETA


soma_produtos = 0
soma_quadrados = 0

for i in range(len(lista1)):
    soma_produtos += (lista1[i] - media1) * (lista2[i] - media2)
    soma_quadrados += (lista1[i] - media1) ** 2

beta = soma_produtos / soma_quadrados
print(f'\nO coeficiente beta da equação de regressão linear simples é: {beta:.2f}')


# 16. COEFIENTE A OU ALPHA
# a = Σ((xi - x̄) * (yi - ȳ)) / Σ((xi - x̄)^2)

#xi e yi são os valores das variáveis independentes e dependentes, respectivamente.
# x̄ e ȳ são as médias das variáveis independentes e dependentes, respectivamente.

media1 = sum(lista1) / len(lista1)
media2 = sum(lista2) / len(lista2)

a = ((media2) - (media1)*beta)

print(f'\nO coeficiente de regressão linear (a ou alpha) é: {a:.2f}')





```
